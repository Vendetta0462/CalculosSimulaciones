# Efecto producido por una onda sobre un anillo de particulas

```python
import numpy as np
import matplotlib.pyplot as plt

def modificar_anillo(N, R, A, wt):
    """
    Genera N puntos en un crculo de radio R y modifica sus radios 
    aplicando el desplazamiento deltaR = R*A*sin(w*t)*(sin(2*theta)+cos(2*theta)).
    
    Parmetros:
      N : int    - Nmero de puntos.
      R : float  - Radio original.
      A : float  - Amplitud del desplazamiento.
      w : float  - Frecuencia angular.
      t : float  - Tiempo o fase.
      
    Retorna:
      x_mod, y_mod : coordenadas de los puntos modificados.
    """
    # Generar los ngulos uniformemente distribuidos entre 0 y 2*pi
    theta = np.linspace(0, 2*np.pi, N, endpoint=False)
    
    # Calcular el desplazamiento radial para cada punto
    delta_R = R * A * np.sin(wt) * (np.sin(2*theta) + np.cos(2*theta))
    
    # Nuevo radio de cada punto
    r_mod = R + delta_R
    
    # Convertir a coordenadas cartesianas
    x_mod = r_mod * np.cos(theta)
    y_mod = r_mod * np.sin(theta)
    
    return x_mod, y_mod

def graficar_anillo(x, y, R):
    """
    Grafica los puntos modificados y el crculo original para comparacin.
    """
    plt.figure(figsize=(7,7))
    
    # Graficar los puntos modificados
    plt.plot(x, y, 'o', label='Puntos modificados', markersize=6)
    
    # Graficar el crculo original (anillo base) para referencia
    theta_c = np.linspace(0, 2*np.pi, 400)
    x_c = R * np.cos(theta_c)
    y_c = R * np.sin(theta_c)
    plt.plot(x_c, y_c, '--', color='gray', label='Crculo original')
    
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Anillo de puntos con desplazamiento radial')
    plt.axis('equal')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    # Parmetros
    N = 200       # Nmero de puntos
    R = 1.0       # Radio original del crculo
    A = 0.3       # Amplitud del desplazamiento
    wt = np.pi/6       # Fase del seno
    
    # Obtener las coordenadas modificadas
    x_mod, y_mod = modificar_anillo(N, R, A, wt)
    
    # Graficar el resultado
    graficar_anillo(x_mod, y_mod, R)
```
